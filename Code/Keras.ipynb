```
import numpy as np 
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras import optimizers
import matplotlib
import matplotlib.pyplot as plt

# 1. create pseudo data y = 2*x0 + 3*x1 + 4
X = np.random.rand(1000, 1)
y = 4 + 3 * X + .2*np.random.randn(1000, 1) # noise added

# 2. Build model 
model = Sequential([Dense(1, input_shape = (1,), activation='linear')])

# 3. gradient descent optimizer and loss function 
sgd = optimizers.SGD(lr=0.1)
model.compile(loss='mse', optimizer=sgd)

# 4. Train the model 
model.fit(X, y, epochs=100, batch_size=2)
```

```
w_gd = model.get_weights()
w_gd_0 = w_gd[0][0] # choose coordinates cho w_1_0
w_gd_1 = w_gd[1][0] # choose coordinates cho w_1_1
x_gd_0 = np.linspace(0, 1, 2, endpoint=True) 
y_gd_0 = w_gd_0 + w_gd_1*x_gd_0 

# Draw the fitting line 
plt.plot(X.T, y.T, 'b.')     # data 
plt.plot(x_gd_0, y_gd_0, 'y', linewidth = 2)   # the fitting line
plt.axis([0, 1, 0, 10])
plt.show()
```
