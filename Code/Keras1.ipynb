```
import numpy as np 
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras import optimizers
import matplotlib
import matplotlib.pyplot as plt

# 1. create pseudo data y = 3 * X + 4
X = np.random.rand(1000, 1)
y = 4 + 3 * X + .2*np.random.randn(1000, 1) # noise added

# 2. Build model 
model = Sequential([Dense(1, input_shape = (1,), activation='linear')]) #Build a model with 1-unit output layer, 1-unit input layer (1 dimension vector), activation is linear function

# 3. gradient descent optimizer and loss function 
sgd = optimizers.SGD(lr=0.1) #choose algorithm SGD, learning rate = 0.1
model.compile(loss='mse', optimizer=sgd) # loss function

# 4. Train the model 
model.fit(X, y, epochs=100, batch_size=2)
```

```
model.fit(X, y, epochs=100, batch_size=2) #epoch= vòng lặp, giới hạn vòng lặp = 100, kích thước mini-batch = 2
```

```
# Get w 
model.get_weights()
```

```
# Display:
w_gd = model.get_weights()
w_gd_0 = w_gd[0][0] 
w_gd_1 = w_gd[1][0] 
x_gd_0 = np.linspace(0, 1, 2, endpoint=True) 
y_gd_0 = w_gd_0 + w_gd_1*x_gd_0 

# Draw the fitting line 
plt.plot(X.T, y.T, 'b.')     # data 
plt.plot(x_gd_0, y_gd_0, 'y', linewidth = 2)   # the fitting line
plt.axis([0, 1, 0, 10])
plt.show()
```
